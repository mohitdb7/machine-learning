```python
import pandas as pd
import numpy as numpy

# vectorizing the sentences; removing stop words
from sklearn.feature_extraction.text import CountVectorizer
```

```python
sample_data = np.array([
    ["Coffee Tea  Soup Coffee Coffee", "Hot"],
    ["Coffee is hot and so is Soup  and Tea", "Hot"],
    ["Espresso is a hot Coffee  and not a Tea", "Hot"],
    ["Coffee is neither Tea nor Soup", "Hot"],
    ["Sprite Pepsi  Cold Coffee and cold Tea", "Cold"]
])
```

```python
hotCold_df = pd.DataFrame(sample_data, columns=["Document", "Class"])
hotCold_df.head()
```

```python
hotCold_df_array = hotCold_df.values
hotCold_X_axis = hotCold_df_array[:,0]
hotCold_y_axis = hotCold_df_array[:,1]

vect = CountVectorizer(stop_words="english")
vect.fit(hotCold_X_axis)
print(len(vect.vocabulary_), vect.vocabulary_)
```

```python
hotCold_X_transformed = vect.transform(hotCold_X_axis)
hotCold_Dataframe_wordBag = pd.DataFrame(hotCold_X_transformed.toarray(), columns=vect.get_feature_names_out())
hotCold_Dataframe_wordBag
```

```python
new_vect = CountVectorizer(stop_words="english")
new_vect.fit(["I hate cold Coffee but love Tea and hot Coffee"])
new_vect.vocabulary_
```
